```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])

df.head
```

```python
X = df.iloc[:,2:]
X
```

```python
y = df['diagnosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(X_train)
X_train = mm.transform(X_train)
X_test = mm.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential    #or from tf.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann = tf.keras.Sequential()
```

```python
#input layer

ann.add(Dense(units = 30,activation = 'relu'))   #dense - all neurons in the initial layer should connected with the neurons of next layer
```

```python
#first hidden layer

ann.add(Dense(units = 8,activation = 'relu'))
```

```python
#second hidden layer

ann.add(Dense(units = 6,activation = 'relu'))
```

```python
#third hidden layer

ann.add(Dense(units = 4,activation = 'relu'))
```

```python
#output layer

ann.add(Dense(units = 1,activation = 'sigmoid'))
```

```python
ann.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
                                                #to update weights --optimizer
                                                #for binary classification - to calculate loss fn --loss
                                                #to evaluate performance measure --metrics

```

```python
#fitting

ann.fit(X_train,y_train,batch_size = 32,epochs = 100)  #epoch --repeated training...learning train data 100 times repeatedly
                                                       #batch_size --train data split to 32 batches each time 32 data obs are trained
                                                       #1-32   32-64 ..so on
```

```python
ann.summary()  #248 = (30(neurons)+1(bias)))*8(first hidden layer neurons) ie, 31*8 = 248
               #54 = (8+1)*6
               #28 = (6+1)*4
               #5 = (4+1)*1    
               #930 = 30*31
```

```python
y_pred = ann.predict(X_test)
print(y_pred)   #gives probabilities of getting 0 and 1

y_pred = (y_pred > 0.5)
print(y_pred)
```

```python
y_test
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
